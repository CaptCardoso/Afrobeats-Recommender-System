This function extracts the uri from the playlists url

```python
def get_uri(url):
    return url.split("/")[-1].split("?")[0]
```

#### Using the Afrobeats Top 100 playlist 

```python
afrobeats_top_100 = "https://open.spotify.com/playlist/32HovauslOCYTENgeZlk6M"
```

```python
playlist_uri = get_uri(afrobeats_top_100)
```

```python
#get json info form the playlist
tracks = sp.playlist_tracks(playlist_uri)
```

This function gets required info from the json (tracks)

```python
afrobeat_dict = {
    'track_uri':[],
    'track_name':[],
    'artist_name':[],
    'artist_info':[],
    'artist_uri':[],
    'artist_popularity':[],
    'artist_genre':[],
    'album':[],
    'track_pop':[],
}

for track in tracks["items"]:
    #URI
    afrobeat_dict['track_uri'].append((track["track"]["uri"]).split(':')[2])
    
    #Track name
    afrobeat_dict['track_name'].append(track["track"]["name"])
    
    #Main Artist
    afrobeat_dict['artist_uri'].append((track["track"]["artists"][0]["uri"]).split(':')[2])
    afrobeat_dict['artist_info'].append(sp.artist(track["track"]["artists"][0]["uri"]))
    
    #Name, popularity, genre
    afrobeat_dict['artist_name'].append(track["track"]["artists"][0]["name"])
    afrobeat_dict['artist_popularity'].append(sp.artist(track["track"]["artists"][0]["uri"])["popularity"])
    afrobeat_dict['artist_genre'].append(sp.artist(track["track"]["artists"][0]["uri"])["genres"])
    
    #Album
    afrobeat_dict['album'].append(track["track"]["album"]["name"])
    
    #Popularity of the track
    afrobeat_dict['track_pop'].append(track["track"]["popularity"])
    time.sleep(1)
```
